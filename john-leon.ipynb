# Question #1

```python
# Partial Question 1 : Resolved

q1 = QuantumRegister(6,'q')
c1 = ClassicalRegister(6,'c')
cc = QuantumRegister(1,'scratch') # custom circuit

QCircuit = QuantumCircuit(q1,cc,c1)

QCircuit.h(q1)
QCircuit.x(cc)
QCircuit.h(cc)
QCircuit.cx(0,6)
QCircuit.cx(1,6)
QCircuit.cx(2,6)
QCircuit.cx(3,6)
QCircuit.cx(4,6)
QCircuit.cx(5,6)

QCircuit.barrier(q1)

QCircuit.h(q1)

QCircuit.measure(0,0)
QCircuit.measure(1,1)
QCircuit.measure(2,2)
QCircuit.measure(3,3)
QCircuit.measure(4,4)
QCircuit.measure(5,5)

QCircuit.draw()
```
